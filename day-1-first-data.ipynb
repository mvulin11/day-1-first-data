```python
import yfinance as yf
import matplotlib.pyplot as plt
apple = yf.download('AAPL', start = '2023-01-01' , end = '2024-01-01')
apple['Close'] .plot(title= 'Apple Stock Price')
plt.show()
```

```python
# Get multiple stocks
stocks = ['AAPL', 'GOOGl', 'MSFT',]
data = yf.download(stocks, start= '2023-01-01', end= '2024-01-01')

#Plot closing prices
data['Close'] .plot(title = 'Stock Comparison', figsize=(10,6))
plt.show()
       
```
